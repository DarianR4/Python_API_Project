```python
 # Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os 
from wconfig import g_key
import requests
import json
```

```python
weatherpy_df = pd.read_csv('WeatherPy_Database.csv')
weatherpy_df
```

```python
#  Humidity Heatmap
# Configure gmaps
gmaps.configure(api_key=g_key)
# Use the Lat and Lng as locations and Humidity as the weight.
locations = weatherpy_df[["Latitudes", "Longitudes"]].astype(float)

# weatherpy_df = weatherpy_df.dropna()
humidity = weatherpy_df['Humidity'].astype(float)

# # Add Heatmap layer to map.
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 1)

fig.add_layer(heat_layer)

fig
```

```python
# set up my ideal weather coordinates.... the aim was to get under 20 results

ideal_weather_df = weatherpy_df.drop(weatherpy_df[(weatherpy_df['Max Temp'] <=72) | (weatherpy_df['Humidity'] >=63)| (weatherpy_df['Wind Speed'] >=3.99) | (weatherpy_df['Cloudiness'] >=30) ].index)

ideal_weather_df = ideal_weather_df.dropna()

ideal_weather_df

```

```python
# Adding the hotel column
hotel_df=ideal_weather_df

hotel_df["Hotel Name"] = ""
hotel_df["City"] = ""


hotel_df
```

```python
# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# set up a parameters dictionary
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key}

for index, row in hotel_df.iterrows():   
    
    latitude = row["Latitudes"]
    longitude = row["Longitudes"]
    params["location"] = f"{latitude},{longitude}"

    # run a request using our params dictionary
    response = requests.get(base_url, params=params).json()
 
    try:
        #     get results
        results = response['results']
        hotel_df.loc[index,"Hotel Name"] = results[0]['name']
        hotel_df.loc[index,"City"] = results[0]['vicinity']
    
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
    
```
